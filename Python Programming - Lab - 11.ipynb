### 01) WAP to demonstrate the use of Pie chart.

```python
x=[10,20,30,40,50]
y=["food","travel","Accomodation","Shoping","Misc"]
c=['r','g','m','y','b']
ex=[0.5,0.0,0.0,0.0,0.0]
plt.title("Trip")
plt.pie(x,labels=y,colors=c,explode=ex,shadow=True,autopct='%1.4f%%',labeldistance=1.2,textprops={"fontsize":11,"color":'yellow'})
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x=["python","c","c++","Java"]
y=[85,70,60,82]
def addlabel(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
c=['r','g','m','y']
plt.bar(x,y,color=c,align='edge',edgecolor='b',alpha=0.7)
plt.title("popularity of language")
plt.xlabel("language")
plt.ylabel("poularity")
addlabel(x,y)
plt.show()

    
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
c=['r','g','m','y','c']
size=[100,200,300,400,500]
plt.scatter(x,y,color=c,marker='*',s=size,edgecolor="blue")
plt.xlabel("id")
plt.ylabel("scatter")
plt.show()
```
